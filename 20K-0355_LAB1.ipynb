```python
#Q7

image = cv2.imread(r'F:\Uni Work\CV Lab\Lab 1\image.jpg')

pixel_values = gray.reshape(-1)
gray_df = pd.DataFrame({"Pixel Values": pixel_values})

print("Basic Statistics of Image Pixel Values:")
print(gray_df.describe())



matrix = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), 60, 1)
rotated = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

cv2.imshow('ROTATED', rotated)
cv2.waitKey(0)
```

```python
#Q8

import matplotlib.pyplot as plt

image2 = cv2.imread(r'F:\Uni Work\CV Lab\Lab 1\image2.jpg')
image2 = cv2.resize(image2, (image.shape[1], image.shape[0]))

alpha = 0.5
beta = 0.5
gamma = 0

blend = cv2.addWeighted(image, alpha, image2, beta, gamma)
gray_blend = cv2.cvtColor(blend, cv2.COLOR_BGR2GRAY)


equal = cv2.equalizeHist(gray_blend)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB))


plt.title("Blended")

plt.subplot(1, 2, 2)

plt.imshow(equal, cmap = "gray") 
plt.title("Enhanced")

plt.show()
```

```python
#Q9

height, width, _ = image.shape
mask = np.zeros((height, width), dtype=np.uint8)

cv2.rectangle(mask, (200, 150), (500, 400), 255, -1)

bitwise_and = cv2.bitwise_and(image, image, mask=mask) 
bitwise_or = cv2.bitwise_or(image, image, mask=mask) 
bitwise_xor = cv2.bitwise_xor(image, image, mask=mask) 
bitwise_not = cv2.bitwise_not(image, mask=mask)

plt.figure(figsize=(12, 8)) 
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 

plt.subplot(2, 3, 2) 
plt.imshow(mask, cmap='gray')
plt.title("mask") 

plt.subplot(2, 3, 3) 
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB)) 
plt.title("and")

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB)) 
plt.title("or")

plt.subplot(2, 3, 6) 
plt.imshow(cv2.cvtColor(bitwise_not, cv2.COLOR_BGR2RGB)) 
plt.title("not")

plt.subplot(2, 3, 5) 
plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB)) 
plt.title("xor")

plt.show()
```

```python
#Q10
image = cv2.imread(r'F:\Uni Work\CV Lab\Lab 1\image.jpg')

statistics = gray_df.describe()

height, width, _ = image.shape

mask = np.zeros((height, width), dtype=np.uint8)

cv2.rectangle(mask, (200, 100), (350, 300), 255, -1) 

masked = cv2.bitwise_and(image, image, mask=mask)

plt.figure(figsize = (14, 8)) 
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB)) 

plt.subplot(1, 3, 3)
plt.table(cellText=statistics.values, colLabels = statistics.columns, cellLoc = 'center', loc = 'center')
plt.show()
```
