```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split

size = (224, 224)
imgs = []
labels = []

for name in os.listdir(loc):
    loc_class = os.path.join(loc, name)
    
    if os.path.isdir(loc_class):
        for name in os.listdir(loc_class):
            img_loc = os.path.join(loc_class, name)
            
            img = cv2.imread(img_loc)
            img = cv2.resize(img, size)  # Use 'img' instead of 'image'
            img = img.astype(np.float32) / 255
            
            imgs.append(img)
            labels.append(name)

imgs = np.array(imgs)
labels = np.array(labels)

X_train, X_test, y_train, y_test = train_test_split(imgs, labels, test_size=0.2, random_state=1)
```
